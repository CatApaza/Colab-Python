```python
#1. Sistema de vehículos
class Vehiculo:
    def describir(self):
        pass

class Auto(Vehiculo):
    def describir(self):
        return "Soy un auto: rápido y cómodo."

class Moto(Vehiculo):
    def describir(self):
        return "Soy una moto: ágil y económica."

class Camion(Vehiculo):
    def describir(self):
        return "Soy un camión: fuerte y resistente."

vehiculos = [Auto(), Moto(), Camion()]

for v in vehiculos:
    print(v.describir())
```

```python
# 2. Inventario de productos con descuento
class Producto:
    def __init__(self, nombre, precio, cantidad):
        self.nombre = nombre
        self.precio = precio
        self.cantidad = cantidad

    def valor_total(self):
        return self.precio * self.cantidad

class ProductoPerecedero(Producto):
    def __init__(self, nombre, precio, cantidad, dias_caducar):
        super().__init__(nombre, precio, cantidad)
        self.dias_caducar = dias_caducar

class ProductoNoPerecedero(Producto):
    def __init__(self, nombre, precio, cantidad, tipo):
        super().__init__(nombre, precio, cantidad)
        self.tipo = tipo

inventario = [
    ProductoPerecedero("Leche", 3.5, 12, 5),
    ProductoNoPerecedero("Arroz", 2.0, 8, "Grano"),
    Producto("Jabón", 1.5, 15)
]

total = 0

for p in inventario:
    subtotal = p.valor_total()
    if p.cantidad > 10:
        subtotal *= 0.9
    print(f"{p.nombre}: ${subtotal:.2f}")
    total += subtotal

print(f"Valor total del inventario: ${total:.2f}")
```

```python
#3. Sistema bancario
class CuentaBancaria:
    def __init__(self, titular, saldo=0):
        self.titular = titular
        self.saldo = saldo

    def depositar(self, monto):
        if monto > 0:
            self.saldo += monto
            print(" Depósito exitoso.")
        else:
            print(" Monto inválido.")

    def retirar(self, monto):
        if monto > 0 and monto <= self.saldo:
            self.saldo -= monto
            print("Retiro exitoso.")
        else:
            print("Fondos insuficientes o monto inválido.")

    def mostrar_tipo(self):
        return "Cuenta Bancaria"

    def consultar_saldo(self):
        print(f"Saldo actual: ${self.saldo:.2f}")

class CuentaAhorro(CuentaBancaria):
    def mostrar_tipo(self):
        return "Cuenta de Ahorro"

class CuentaCorriente(CuentaBancaria):
    def mostrar_tipo(self):
        return "Cuenta Corriente"

cuenta = CuentaAhorro("Ana", 1000)

while True:
    print(f"\n--- {cuenta.mostrar_tipo()} de {cuenta.titular} ---")
    print("1. Consultar saldo")
    print("2. Depositar")
    print("3. Retirar")
    print("4. Salir")
    opcion = input("Elige una opción: ")

    if opcion == "1":
        cuenta.consultar_saldo()
    elif opcion == "2":
        try:
            monto = float(input("Monto a depositar: "))
            cuenta.depositar(monto)
        except ValueError:
            print("Ingresa un número válido.")
    elif opcion == "3":
        try:
            monto = float(input("Monto a retirar: "))
            cuenta.retirar(monto)
        except ValueError:
            print(" Ingresa un número válido.")
    elif opcion == "4":
        print("¡Gracias por usar el sistema bancario!")
        break
    else:
        print(" Opción no válida.")
```

```python
#4. Gestión de empleados
class Empleado:
    def __init__(self, nombre, salario_base):
        self.nombre = nombre
        self.salario_base = salario_base

    def calcular_salario(self):
        return self.salario_base

    def mostrar_tipo(self):
        return "Empleado General"

class EmpleadoTiempoCompleto(Empleado):
    def calcular_salario(self):
        bonificacion = 0.20 * self.salario_base
        return self.salario_base + bonificacion

    def mostrar_tipo(self):
        return "Empleado a Tiempo Completo"

class EmpleadoMedioTiempo(Empleado):
    def calcular_salario(self):
        bonificacion = 0.10 * self.salario_base
        return self.salario_base + bonificacion

    def mostrar_tipo(self):
        return "Empleado Medio Tiempo"

empleados = [
    EmpleadoTiempoCompleto("Laura", 2000),
    EmpleadoMedioTiempo("Carlos", 1000),
    EmpleadoTiempoCompleto("Marta", 2500),
    EmpleadoMedioTiempo("Luis", 1200)
]

print("Lista de empleados y sus salarios:")
for empleado in empleados:
    tipo = empleado.mostrar_tipo()
    salario = empleado.calcular_salario()
    print(f"{tipo} - {empleado.nombre}: ${salario:.2f}")
```

```python
#5. Figuras geométricas
import math

class Figura:
    def area(self):
        return 0

    def nombre(self):
        return "Figura genérica"

class Rectangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        return self.base * self.altura

    def nombre(self):
        return "Rectángulo"

class Triangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        return (self.base * self.altura) / 2

    def nombre(self):
        return "Triángulo"

class Circulo(Figura):
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return math.pi * self.radio ** 2

    def nombre(self):
        return "Círculo"


figuras = [
    Rectangulo(4, 5),
    Triangulo(6, 3),
    Circulo(2.5),
    Rectangulo(2, 8),
    Circulo(1)
]


print("Áreas de figuras geométricas:")
for figura in figuras:
    print(f"{figura.nombre()} - Área: {figura.area():.2f}")
```

```python
# 7. Registro de estudiantes y notas
class Estudiante:
    def __init__(self, nombre, notas):
        self.nombre = nombre
        self.notas = notas

    def promedio(self):
        if len(self.notas) == 0:
            return 0
        return sum(self.notas) / len(self.notas)

    def aprobado(self):
        return self.promedio() >= 11

    def tipo(self):
        return "Estudiante regular"

class EstudianteBecado(Estudiante):
    def aprobado(self):
        return self.promedio() >= 14

    def tipo(self):
        return "Estudiante becado"


estudiantes = [
    Estudiante("Luis", [12, 10, 11]),
    EstudianteBecado("María", [15, 14, 13]),
    Estudiante("Javier", [8, 9, 10]),
    EstudianteBecado("Ana", [13, 13, 14])
]


print("Lista de estudiantes y su estado:")
for estudiante in estudiantes:
    estado = "Aprobado" if estudiante.aprobado() else "Desaprobado"
    print(f"{estudiante.tipo()} - {estudiante.nombre} | Promedio: {estudiante.promedio():.2f} | {estado}")
```

```python
#10. Agenda de contactos
class Contacto:
    def __init__(self, nombre, telefono):
        self.nombre = nombre
        self.telefono = telefono

    def saludo(self):
        return f"Hola, soy {self.nombre}."

class Amigo(Contacto):
    def saludo(self):
        return f"¡Hola, amigo {self.nombre}! ¿Cómo estás?"

class Familiar(Contacto):
    def saludo(self):
        return f"¡Hola, {self.nombre}! ¿Cómo está la familia?"

class Trabajo(Contacto):
    def saludo(self):
        return f"Hola {self.nombre}, ¿cómo van los proyectos?"


contactos = []


while True:
    print("\n Menú - Agenda de contactos")
    print("1. Agregar contacto")
    print("2. Buscar contacto")
    print("3. Salir")
    opcion = input("Elige una opción: ")

    if opcion == "1":
        nombre = input("Nombre del contacto: ")
        telefono = input("Teléfono del contacto: ")
        print("Tipo de contacto: 1. Amigo  2. Familiar  3. Trabajo")
        tipo = input("Selecciona tipo de contacto: ")

        if tipo == "1":
            contactos.append(Amigo(nombre, telefono))
        elif tipo == "2":
            contactos.append(Familiar(nombre, telefono))
        elif tipo == "3":
            contactos.append(Trabajo(nombre, telefono))
        else:
            print("Tipo no válido.")
    elif opcion == "2":
        buscar_nombre = input("Introduce el nombre del contacto a buscar: ")
        encontrado = False
        for contacto in contactos:
            if contacto.nombre.lower() == buscar_nombre.lower():
                print(f"{contacto.saludo()} | Teléfono: {contacto.telefono}")
                encontrado = True
                break
        if not encontrado:
            print(" Contacto no encontrado.")
    elif opcion == "3":
        print(" ¡Gracias por usar la agenda de contactos!")
        break
    else:
        print(" Opción no válida. Intenta de nuevo.")
```

```python
#11: Figuras geométricas
import math

class Figura:
    def area(self):
        pass

    def perimetro(self):
        pass

class Rectangulo(Figura):
    def __init__(self, largo, ancho):
        self.largo = largo
        self.ancho = ancho

    def area(self):
        return self.largo * self.ancho

    def perimetro(self):
        return 2 * (self.largo + self.ancho)

class Circulo(Figura):
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return math.pi * self.radio ** 2

    def perimetro(self):
        return 2 * math.pi * self.radio

figuras = [
    Rectangulo(5, 3),
    Circulo(7),
    Rectangulo(10, 4),
    Circulo(3)
]

for figura in figuras:
    print(f"\n{figura.__class__.__name__} - Área: {figura.area():.2f} | Perímetro: {figura.perimetro():.2f}")
```

```python
# 14: Inventario de una tienda
class Producto:
    def __init__(self, nombre, precio, stock):
        self.nombre = nombre
        self.precio = precio
        self.stock = stock

    def aplicar_descuento(self):
        return self.precio

class Ropa(Producto):
    def aplicar_descuento(self):
        if self.stock > 10:
            return self.precio * 0.9
        return self.precio

class Alimento(Producto):
    def aplicar_descuento(self):
        if self.stock > 20:
            return self.precio * 0.85
        return self.precio

class Tecnologia(Producto):
    def aplicar_descuento(self):
        if self.stock > 5:
            return self.precio * 0.95
        return self.precio

productos = [
    Ropa("Polo", 50, 12),
    Alimento("Arroz", 80, 25),
    Tecnologia("Laptop", 3000, 2)
]

for producto in productos:
    print(f"{producto.nombre} - Precio con descuento: {producto.aplicar_descuento()}")

for producto in productos:
    while producto.stock < 5:
        print(f"¡Advertencia! Bajo stock en {producto.nombre}: {producto.stock}")
        producto.stock += 1
```

```python
#15: Sistema escolar
class Persona:
    def __init__(self, nombre):
        self.nombre = nombre

    def rol(self):
        return "Persona"

class Estudiante(Persona):
    def __init__(self, nombre):
        super().__init__(nombre)
        self.notas = []

    def agregar_nota(self, nota):
        self.notas.append(nota)

    def calcular_promedio(self):
        return sum(self.notas) / len(self.notas) if self.notas else 0

    def rol(self):
        return "Estudiante - Aprende"

class Profesor(Persona):
    def rol(self):
        return "Profesor - Enseña"

est1 = Estudiante("Lucía")
est2 = Estudiante("Marco")
prof = Profesor("Tania")

est1.agregar_nota(18)
est1.agregar_nota(16)
est2.agregar_nota(10)
est2.agregar_nota(12)

personas = [est1, est2, prof]

for persona in personas:
    print(f"{persona.nombre}: {persona.rol()}")
    if isinstance(persona, Estudiante):
        promedio = persona.calcular_promedio()
        print(f"  Promedio: {promedio:.2f}")
        if promedio >= 13:
            print("  Estado: Aprobado")
        else:
            print("  Estado: Desaprobado")
```

```python
#16: Cajero automático
class CuentaBancaria:
    def __init__(self, titular, saldo=0):
        self.titular = titular
        self.saldo = saldo

    def depositar(self, monto):
        self.saldo += monto
        print(f"Depósito exitoso. Saldo actual: {self.saldo}")

    def retirar(self, monto):
        if monto <= self.saldo:
            self.saldo -= monto
            print(f"Retiro exitoso. Saldo actual: {self.saldo}")
        else:
            print("Fondos insuficientes.")

    def consultar_saldo(self):
        print(f"Saldo actual: {self.saldo}")

class CuentaAhorro(CuentaBancaria):
    def __init__(self, titular, saldo=0, interes=0.02):
        super().__init__(titular, saldo)
        self.interes = interes

    def aplicar_interes(self):
        self.saldo *= (1 + self.interes)
        print(f"Interés aplicado. Nuevo saldo: {self.saldo}")

class CuentaCorriente(CuentaBancaria):
    def __init__(self, titular, saldo=0, sobregiro=500):
        super().__init__(titular, saldo)
        self.sobregiro = sobregiro

    def retirar(self, monto):
        if monto <= (self.saldo + self.sobregiro):
            self.saldo -= monto
            print(f"Retiro exitoso. Saldo actual: {self.saldo}")
        else:
            print("Límite de sobregiro excedido.")

cuenta = CuentaAhorro("Juan Pérez", 1000)

while True:
    print("\n--- Menú Cajero ---")
    print("1. Consultar saldo")
    print("2. Depositar")
    print("3. Retirar")
    print("4. Aplicar interés (Ahorro)")
    print("5. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        cuenta.consultar_saldo()
    elif opcion == "2":
        monto = float(input("Ingrese monto a depositar: "))
        cuenta.depositar(monto)
    elif opcion == "3":
        monto = float(input("Ingrese monto a retirar: "))
        cuenta.retirar(monto)
    elif opcion == "4" and isinstance(cuenta, CuentaAhorro):
        cuenta.aplicar_interes()
    elif opcion == "5":
        print("Gracias por usar nuestro servicio.")
        break
    else:
        print("Opción no válida.")
```

```python
# 18: Control de empleados
class Empleado:
    def __init__(self, nombre, salario):
        self.nombre = nombre
        self.salario = salario

    def calcular_bono(self):
        return self.salario * 0.10

class Gerente(Empleado):
    def calcular_bono(self):
        return self.salario * 0.20 + 1000

class Asistente(Empleado):
    def calcular_bono(self):
        return self.salario * 0.05 + 500

empleados = [
    Empleado("Juan", 30000),
    Gerente("María", 50000),
    Asistente("Pedro", 25000),
    Gerente("Ana", 55000),
    Asistente("Luisa", 22000)
]

for empleado in empleados:
    bono = empleado.calcular_bono()

    if bono > 5000:
        print(f"{empleado.nombre} (Gerente) recibe un bono especial de {bono:.2f}")
    elif bono > 1000:
        print(f"{empleado.nombre} recibe un bono estándar de {bono:.2f}")
    else:
        print(f"{empleado.nombre} recibe un bono básico de {bono:.2f}")
```

```python
#19: Lotería
import random

class Boleto:
    def __init__(self, numero):
        self.numero = numero
        self.premio = None

    def asignar_premio(self):
        premios = ["Nada", "100", "500", "1000", "5000"]
        self.premio = random.choice(premios)

class Loteria:
    def __init__(self):
        self.boletos = []

    def generar_boletos(self, cantidad):
        for i in range(cantidad):
            boleto = Boleto(i + 1)
            boleto.asignar_premio()
            self.boletos.append(boleto)

    def mostrar_boletos(self):
        for boleto in self.boletos:
            print(f"Boleto #{boleto.numero} - Premio: {boleto.premio}")

    def detectar_ganadores(self):
        for boleto in self.boletos:
            if boleto.premio != "Nada":
                print(f"¡Boleto #{boleto.numero} ha ganado {boleto.premio}!")
            else:
                print(f"Boleto #{boleto.numero} no tiene premio.")

loteria = Loteria()
loteria.generar_boletos(5)
loteria.mostrar_boletos()
print("\nDetectando ganadores...")
loteria.detectar_ganadores()
```

```python
#20: Biblioteca
class Material:
    def tipo_material(self):
        pass

class Libro(Material):
    def tipo_material(self):
        return "Libro"

class Revista(Material):
    def tipo_material(self):
        return "Revista"

class Tesis(Material):
    def tipo_material(self):
        return "Tesis"

materiales = [
    Libro(),
    Revista(),
    Tesis(),
    Tesis(),
    Libro(),
    Libro(),
    Revista(),
    Tesis(),
    Libro(),
    Revista()
]

contador = {"Libro": 0, "Revista": 0, "Tesis": 0}

for material in materiales:
    tipo = material.tipo_material()
    if tipo in contador:
        contador[tipo] += 1
    else:
        contador[tipo] = 1

print("Inventario de la biblioteca Tania:")
for tipo, cantidad in contador.items():
    print(f"{tipo}: {cantidad}")
```
