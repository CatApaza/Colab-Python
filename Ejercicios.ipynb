```python
#1. Sistema de vehículos
class Vehiculo:
    def describir(self):
        pass

class Auto(Vehiculo):
    def describir(self):
        return "Soy un auto: rápido y cómodo."

class Moto(Vehiculo):
    def describir(self):
        return "Soy una moto: ágil y económica."

class Camion(Vehiculo):
    def describir(self):
        return "Soy un camión: fuerte y resistente."

vehiculos = [Auto(), Moto(), Camion()]

for v in vehiculos:
    print(v.describir())
```

```python
# 2. Inventario de productos con descuento
class Producto:
    def __init__(self, nombre, precio, cantidad):
        self.nombre = nombre
        self.precio = precio
        self.cantidad = cantidad

    def valor_total(self):
        return self.precio * self.cantidad

class ProductoPerecedero(Producto):
    def __init__(self, nombre, precio, cantidad, dias_caducar):
        super().__init__(nombre, precio, cantidad)
        self.dias_caducar = dias_caducar

class ProductoNoPerecedero(Producto):
    def __init__(self, nombre, precio, cantidad, tipo):
        super().__init__(nombre, precio, cantidad)
        self.tipo = tipo

inventario = [
    ProductoPerecedero("Leche", 3.5, 12, 5),
    ProductoNoPerecedero("Arroz", 2.0, 8, "Grano"),
    Producto("Jabón", 1.5, 15)
]

total = 0

for p in inventario:
    subtotal = p.valor_total()
    if p.cantidad > 10:
        subtotal *= 0.9  # 10% de descuento
    print(f"{p.nombre}: ${subtotal:.2f}")
    total += subtotal

print(f"Valor total del inventario: ${total:.2f}")
```

```python
#3. Sistema bancario
class CuentaBancaria:
    def __init__(self, titular, saldo=0):
        self.titular = titular
        self.saldo = saldo

    def depositar(self, monto):
        if monto > 0:
            self.saldo += monto
            print(" Depósito exitoso.")
        else:
            print(" Monto inválido.")

    def retirar(self, monto):
        if monto > 0 and monto <= self.saldo:
            self.saldo -= monto
            print("Retiro exitoso.")
        else:
            print("Fondos insuficientes o monto inválido.")

    def mostrar_tipo(self):
        return "Cuenta Bancaria"

    def consultar_saldo(self):
        print(f"Saldo actual: ${self.saldo:.2f}")

class CuentaAhorro(CuentaBancaria):
    def mostrar_tipo(self):
        return "Cuenta de Ahorro"

class CuentaCorriente(CuentaBancaria):
    def mostrar_tipo(self):
        return "Cuenta Corriente"

# Simulación
cuenta = CuentaAhorro("Ana", 1000)

while True:
    print(f"\n--- {cuenta.mostrar_tipo()} de {cuenta.titular} ---")
    print("1. Consultar saldo")
    print("2. Depositar")
    print("3. Retirar")
    print("4. Salir")
    opcion = input("Elige una opción: ")

    if opcion == "1":
        cuenta.consultar_saldo()
    elif opcion == "2":
        try:
            monto = float(input("Monto a depositar: "))
            cuenta.depositar(monto)
        except ValueError:
            print("Ingresa un número válido.")
    elif opcion == "3":
        try:
            monto = float(input("Monto a retirar: "))
            cuenta.retirar(monto)
        except ValueError:
            print(" Ingresa un número válido.")
    elif opcion == "4":
        print("¡Gracias por usar el sistema bancario!")
        break
    else:
        print(" Opción no válida.")
```

```python
#4. Gestión de empleados
class Empleado:
    def __init__(self, nombre, salario_base):
        self.nombre = nombre
        self.salario_base = salario_base

    def calcular_salario(self):
        return self.salario_base

    def mostrar_tipo(self):
        return "Empleado General"

class EmpleadoTiempoCompleto(Empleado):
    def calcular_salario(self):
        bonificacion = 0.20 * self.salario_base
        return self.salario_base + bonificacion

    def mostrar_tipo(self):
        return "Empleado a Tiempo Completo"

class EmpleadoMedioTiempo(Empleado):
    def calcular_salario(self):
        bonificacion = 0.10 * self.salario_base
        return self.salario_base + bonificacion

    def mostrar_tipo(self):
        return "Empleado Medio Tiempo"

# Lista de empleados
empleados = [
    EmpleadoTiempoCompleto("Laura", 2000),
    EmpleadoMedioTiempo("Carlos", 1000),
    EmpleadoTiempoCompleto("Marta", 2500),
    EmpleadoMedioTiempo("Luis", 1200)
]

# Mostrar salarios
print("Lista de empleados y sus salarios:")
for empleado in empleados:
    tipo = empleado.mostrar_tipo()
    salario = empleado.calcular_salario()
    print(f"{tipo} - {empleado.nombre}: ${salario:.2f}")
```

```python
#5. Figuras geométricas
import math

class Figura:
    def area(self):
        return 0

    def nombre(self):
        return "Figura genérica"

class Rectangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        return self.base * self.altura

    def nombre(self):
        return "Rectángulo"

class Triangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def area(self):
        return (self.base * self.altura) / 2

    def nombre(self):
        return "Triángulo"

class Circulo(Figura):
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return math.pi * self.radio ** 2

    def nombre(self):
        return "Círculo"

# Lista de figuras
figuras = [
    Rectangulo(4, 5),
    Triangulo(6, 3),
    Circulo(2.5),
    Rectangulo(2, 8),
    Circulo(1)
]

# Mostrar áreas usando polimorfismo
print("📐 Áreas de figuras geométricas:")
for figura in figuras:
    print(f"{figura.nombre()} - Área: {figura.area():.2f}")
```

```python
#6. Sistema de mascotas
class Mascota:
    def __init__(self, nombre):
        self.nombre = nombre

    def hacer_sonido(self):
        return "Sonido genérico"

    def tipo(self):
        return "Mascota"

class Perro(Mascota):
    def hacer_sonido(self):
        return "Guau guau"

    def tipo(self):
        return "Perro"

class Gato(Mascota):
    def hacer_sonido(self):
        return "Miau"

    def tipo(self):
        return "Gato"

class Pez(Mascota):
    def hacer_sonido(self):
        return "(burbujas)"

    def tipo(self):
        return "Pez"

# Lista para almacenar mascotas
mascotas = []

# Menú interactivo
while True:
    print("\n Menú - Sistema de Mascotas")
    print("1. Agregar mascota")
    print("2. Listar mascotas")
    print("3. Salir")
    opcion = input("Elige una opción: ")

    if opcion == "1":
        nombre = input("Nombre de la mascota: ")
        print("Tipo: 1. Perro  2. Gato  3. Pez")
        tipo = input("Selecciona tipo: ")

        if tipo == "1":
            mascotas.append(Perro(nombre))
        elif tipo == "2":
            mascotas.append(Gato(nombre))
        elif tipo == "3":
            mascotas.append(Pez(nombre))
        else:
            print("Tipo no válido.")
    elif opcion == "2":
        print("\n Lista de mascotas registradas:")
        if not mascotas:
            print("No hay mascotas registradas.")
        for mascota in mascotas:
            print(f"{mascota.tipo()} - {mascota.nombre}: {mascota.hacer_sonido()}")
    elif opcion == "3":
        print("¡Gracias por usar el sistema de mascotas!")
        break
    else:
        print(" Opción no válida. Intenta de nuevo.")
```

```python
# 7. Registro de estudiantes y notas
class Estudiante:
    def __init__(self, nombre, notas):
        self.nombre = nombre
        self.notas = notas

    def promedio(self):
        if len(self.notas) == 0:
            return 0
        return sum(self.notas) / len(self.notas)

    def aprobado(self):
        return self.promedio() >= 11  # Regla general: mínimo 11

    def tipo(self):
        return "Estudiante regular"

class EstudianteBecado(Estudiante):
    def aprobado(self):
        return self.promedio() >= 14  # Regla más estricta

    def tipo(self):
        return "Estudiante becado"

# Lista de estudiantes
estudiantes = [
    Estudiante("Luis", [12, 10, 11]),
    EstudianteBecado("María", [15, 14, 13]),
    Estudiante("Javier", [8, 9, 10]),
    EstudianteBecado("Ana", [13, 13, 14])
]

# Mostrar resultados
print("Lista de estudiantes y su estado:")
for estudiante in estudiantes:
    estado = "Aprobado" if estudiante.aprobado() else "Desaprobado"
    print(f"{estudiante.tipo()} - {estudiante.nombre} | Promedio: {estudiante.promedio():.2f} | {estado}")
```

```python
 #8. Tienda en línea
 class Producto:
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

class Cliente:
    def __init__(self, nombre):
        self.nombre = nombre
        self.carrito = []

    def agregar_producto(self, producto):
        self.carrito.append(producto)

    def calcular_total(self):
        total = sum([producto.precio for producto in self.carrito])
        # Aplicar descuento si hay más de 3 productos
        if len(self.carrito) > 3:
            total *= 0.9  # 10% de descuento
        return total

    def mostrar_carrito(self):
        if not self.carrito:
            print("Tu carrito está vacío.")
        else:
            print("\nProductos en tu carrito:")
            for producto in self.carrito:
                print(f"- {producto.nombre} | ${producto.precio:.2f}")
            print(f"Total: ${self.calcular_total():.2f}")

# Crear algunos productos
producto1 = Producto("Camiseta", 20.0)
producto2 = Producto("Pantalón", 35.0)
producto3 = Producto("Zapatos", 50.0)
producto4 = Producto("Gorra", 10.0)
producto5 = Producto("Chaqueta", 45.0)

# Crear un cliente
cliente = Cliente("Carlos")

# Menú interactivo
while True:
    print("\n Menú de Tienda en línea")
    print("1. Agregar producto al carrito")
    print("2. Ver carrito")
    print("3. Realizar compra")
    print("4. Salir")
    opcion = input("Elige una opción: ")

    if opcion == "1":
        print("\nProductos disponibles:")
        print("1. Camiseta - $20.00")
        print("2. Pantalón - $35.00")
        print("3. Zapatos - $50.00")
        print("4. Gorra - $10.00")
        print("5. Chaqueta - $45.00")
        producto_opcion = input("Elige el número del producto a agregar: ")

        if producto_opcion == "1":
            cliente.agregar_producto(producto1)
        elif producto_opcion == "2":
            cliente.agregar_producto(producto2)
        elif producto_opcion == "3":
            cliente.agregar_producto(producto3)
        elif producto_opcion == "4":
            cliente.agregar_producto(producto4)
        elif producto_opcion == "5":
            cliente.agregar_producto(producto5)
        else:
            print(" Opción no válida.")
    elif opcion == "2":
        cliente.mostrar_carrito()
    elif opcion == "3":
        print(f"\n Compra realizada. Total: ${cliente.calcular_total():.2f}")
        break
    elif opcion == "4":
        print("¡Gracias por visitar nuestra tienda!")
        break
    else:
        print("Opción no válida. Intenta de nuevo.")
```

```python
#9. Zoológico interactivo
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def emitir_sonido(self):
        return "Sonido genérico"

    def tipo(self):
        return "Animal"

class Leon(Animal):
    def emitir_sonido(self):
        return " Ruge"

    def tipo(self):
        return "León"

class Tigre(Animal):
    def emitir_sonido(self):
        return " Ruge fuerte"

    def tipo(self):
        return "Tigre"

class Elefante(Animal):
    def emitir_sonido(self):
        return " Barrita"

    def tipo(self):
        return "Elefante"

# Lista de animales
animales = [
    Leon("Simba"),
    Tigre("Shere Khan"),
    Elefante("Dumbo"),
    Leon("Mufasa"),
    Tigre("Rajah")
]

# Menú interactivo
while True:
    print("\n Menú - Zoológico interactivo")
    print("1. Ver animales")
    print("2. Salir")
    opcion = input("Elige una opción: ")

    if opcion == "1":
        print("\n Detalles del zoológico:")
        for animal in animales:
            print(f"{animal.tipo()} - {animal.nombre}: {animal.emitir_sonido()}")
    elif opcion == "2":
        print(" ¡Gracias por visitar el zoológico!")
        break
    else:
        print("Opción no válida. Intenta de nuevo.")
```

```python
#10. Agenda de contactos
class Contacto:
    def __init__(self, nombre, telefono):
        self.nombre = nombre
        self.telefono = telefono

    def saludo(self):
        return f"Hola, soy {self.nombre}."

class Amigo(Contacto):
    def saludo(self):
        return f"¡Hola, amigo {self.nombre}! ¿Cómo estás?"

class Familiar(Contacto):
    def saludo(self):
        return f"¡Hola, {self.nombre}! ¿Cómo está la familia?"

class Trabajo(Contacto):
    def saludo(self):
        return f"Hola {self.nombre}, ¿cómo van los proyectos?"

# Lista para almacenar contactos
contactos = []

# Menú interactivo
while True:
    print("\n Menú - Agenda de contactos")
    print("1. Agregar contacto")
    print("2. Buscar contacto")
    print("3. Salir")
    opcion = input("Elige una opción: ")

    if opcion == "1":
        nombre = input("Nombre del contacto: ")
        telefono = input("Teléfono del contacto: ")
        print("Tipo de contacto: 1. Amigo  2. Familiar  3. Trabajo")
        tipo = input("Selecciona tipo de contacto: ")

        if tipo == "1":
            contactos.append(Amigo(nombre, telefono))
        elif tipo == "2":
            contactos.append(Familiar(nombre, telefono))
        elif tipo == "3":
            contactos.append(Trabajo(nombre, telefono))
        else:
            print("Tipo no válido.")
    elif opcion == "2":
        buscar_nombre = input("Introduce el nombre del contacto a buscar: ")
        encontrado = False
        for contacto in contactos:
            if contacto.nombre.lower() == buscar_nombre.lower():
                print(f"{contacto.saludo()} | Teléfono: {contacto.telefono}")
                encontrado = True
                break
        if not encontrado:
            print(" Contacto no encontrado.")
    elif opcion == "3":
        print(" ¡Gracias por usar la agenda de contactos!")
        break
    else:
        print(" Opción no válida. Intenta de nuevo.")
```

```python
#11: Figuras geométricas
import math

class Figura:
    def area(self):
        pass

    def perimetro(self):
        pass

class Rectangulo(Figura):
    def __init__(self, largo, ancho):
        self.largo = largo
        self.ancho = ancho

    def area(self):
        return self.largo * self.ancho

    def perimetro(self):
        return 2 * (self.largo + self.ancho)

class Circulo(Figura):
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return math.pi * self.radio ** 2

    def perimetro(self):
        return 2 * math.pi * self.radio

# Lista de figuras
figuras = [
    Rectangulo(5, 3),
    Circulo(7),
    Rectangulo(10, 4),
    Circulo(3)
]

# Mostrar área y perímetro de cada figura
for figura in figuras:
    print(f"\n{figura.__class__.__name__} - Área: {figura.area():.2f} | Perímetro: {figura.perimetro():.2f}")
```

```python
#12: Sistema de usuarios
class Usuario:
    def __init__(self, nombre):
        self.nombre = nombre

    def mostrar_privilegios(self):
        return "Privilegios básicos"

class Administrador(Usuario):
    def mostrar_privilegios(self):
        return "Acceso completo a todas las funciones."

class Cliente(Usuario):
    def mostrar_privilegios(self):
        return "Acceso limitado a funciones de cliente."

# Lista de usuarios
usuarios = [
    Administrador("Carlos"),
    Cliente("Ana"),
    Administrador("Luis"),
    Cliente("Marta")
]

# Mostrar privilegios de cada usuario
for usuario in usuarios:
    print(f"{usuario.nombre} ({usuario.__class__.__name__}) - Privilegios: {usuario.mostrar_privilegios()}")
```
